```python
from csv import reader
import numpy as np

csv_header = np.array([])
csv_data = np.array([])

with open('company_sales_data.csv') as csvfile:
    csvreader = reader(csvfile)

    csv_header = np.append(csv_header, next(csvreader))

    for row in csvreader:
        csv_data = np.append(csv_data, row, axis = 0)


csv_data = csv_data.reshape(-1, 9)

print(csv_header)
print(csv_data)
```

```python
month_numbers = csv_data[:, 0].astype(int)
print(month_numbers)
```

```python
facecream_sales_data = csv_data[:, 1].astype(int)
print(facecream_sales_data)
```

```python
facewash_sales_data = csv_data[:, 2].astype(int)
print(facewash_sales_data)
```

```python
toothpaste_sales_data = csv_data[:, 3].astype(int)
print(toothpaste_sales_data)
```

```python
bathingsoap_sales_data = csv_data[:, 4].astype(int)
print(bathingsoap_sales_data)
```

```python
shampoo_sales_data = csv_data[:, 5].astype(int)
print(shampoo_sales_data)
```

```python
moisturizer_sales_data = csv_data[:, 6].astype(int)
print(moisturizer_sales_data)
```

```python
from matplotlib import pyplot as plt


plt.title('Sales data')
plt.xlabel('Month Number')
plt.ylabel('Sales unit in number')

plt.plot(month_numbers, facecream_sales_data, label = 'Face cream sales data',
        marker = 'o', linewidth = 3)

plt.plot(month_numbers, facewash_sales_data, label = 'Face wash sales data',
        marker = 'o', linewidth = 3)

plt.plot(month_numbers, toothpaste_sales_data, label = 'Toothpaste sales data',
        marker = 'o', linewidth = 3)

plt.plot(month_numbers, bathingsoap_sales_data, label = 'Bathing soap sales data',
        marker = 'o', linewidth = 3)

plt.plot(month_numbers, shampoo_sales_data, label = 'Shampoo sales data', marker = 'o', linewidth = 3)

plt.plot(month_numbers, moisturizer_sales_data, label = 'Moisturizer sales data', marker = 'o', linewidth = 3)

plt.yticks([1000, 2000, 4000, 6000, 8000, 10000, 12000, 15000, 18000])
plt.xticks(month_numbers)

plt.legend(loc = 'upper left')
plt.show()
```

```python
plt.scatter(month_numbers, toothpaste_sales_data, label = 'Tooth paste Sales data')
plt.xlabel('Month Number')
plt.ylabel('Number of units Sold')
plt.title('Toothpaste sales data')
plt.xticks(month_numbers)
plt.grid(linestyle = '--')
plt.legend()

plt.show()
```
