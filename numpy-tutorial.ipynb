```python
# Flatten vs ravel
# Changes made to new array will not affect parent array in case
# of flatten but in ravel, it does.

print(a)

b = a.flatten()
print(b)

b[1] = 20

print(b)
print(a)

b = a.ravel()

print(b)

b[1] = 20

print(b)
print(a)
```
