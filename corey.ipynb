```python
import pandas as pd
from matplotlib import pyplot as plt

data = pd.read_csv('data.csv')
ages = data['Age']
dev_salaries = data['All_Devs']
py_salaries = data['Python']
js_salaries = data['JavaScript']
```

```python
fig, ax = plt.subplots()

ax.plot(ages, dev_salaries, color='#444444', linestyle='--', label='All Devs')

ax.plot(ages, py_salaries, label='Python')

overall_median = 57287

ax.fill_between(ages, py_salaries, dev_salaries,
  where = (py_salaries > dev_salaries),
  alpha = 0.25, interpolate = True, color = 'green', label = 'Above average salary')

ax.fill_between(ages, py_salaries, dev_salaries,
  where = (py_salaries <= dev_salaries),
  alpha = 0.25, interpolate = True, color = 'red', label = 'Below average salary')

plt.fill_between(ages, dev_salaries, alpha = 0.25)

ax.legend()

ax.set_title('Median Salary (USD) by Age')
ax.set_xlabel('Ages')
ax.set_ylabel('Median Salary (USD)')

plt.tight_layout()

plt.show()
```

```python
# fig, ax1 = plt.subplots(nrows = 2, ncols = 1, sharex = True, sharey = True)
fig1, ax1 = plt.subplots()
fig2, ax2 = plt.subplots()

# ax1, ax2 = ax

ax1.plot(ages, js_salaries, label = 'Javascript', color = 'red')
ax1.plot(ages, dev_salaries, label = 'Average', color = 'black', linestyle = '--')
ax1.set_title('Median Salary (USD) by Age')
ax1.set_ylabel('Median Salary')

ax2.plot(ages, py_salaries, label = 'Python', color = 'green')
ax2.plot(ages, dev_salaries, label = 'Average', color = 'black', linestyle = '--')

ax2.set_title('Median Salary (USD) by Age')
ax2.set_xlabel('Age')
ax2.set_ylabel('Median Salary')

ax1.legend()
ax2.legend()

plt.show()

fig1.savefig('fig1.png')
fig2.savefig('fig2.png')
```
